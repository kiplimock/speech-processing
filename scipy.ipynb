```python
from scipy.fft import rfft, rfftfreq
import librosa
from matplotlib import pyplot as plt
import numpy as np

filepath = 'C:\\Users\\kipli\\Cedric\\speech_processing\\go_the_distance.ogg'
data, sr = librosa.load(filepath, sr=None, res_type='fft')

y = rfft(data)
x = rfftfreq(len(data), 1/sr)

plt.figure(figsize=(15,5))
plt.plot(x, np.abs(y))
plt.ylabel("")
plt.xlabel("Frequency (Hz)")
# plt.xlim(0,4000)
plt.show()
# plt.figure()
# plt.plot(list(range(0, len(data))), data)
```
